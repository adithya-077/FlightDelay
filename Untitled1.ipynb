```python
import pandas as pd
import numpy as np
import os
from sklearn.model_selection import train_test_split
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from imblearn.over_sampling import SMOTE
from sklearn.tree import DecisionTreeClassifier
```

```python
os.getcwd()
```

```python
os.chdir(os.getcwd()+'\\Weather dataset')
```

```python
df1 = pd.read_csv('./final_merged.csv')
df1 = df1.drop(['Unnamed: 0'] , axis=1)
df1
```

```python
df1['DelayClass'] = df1['DepDelayMinutes'].apply(lambda x: 1 if x!=0 else 0)
```

```python
 df1.columns
```

```python
df1=df1.dropna()
```

```python
df1 = df1.drop(['FlightDate','Origin','Dest','time','date','airport','DepTime','DepDel15','ArrTime','ArrDel15','ArrDelayMinutes'],axis=1)
```

```python
df1
```

```python
cor= df1.corr() 
cor
```

```python
plt.figure(figsize=(25,25))
sns.heatmap(cor,annot=True)
```

```python
#month and quarter are positivly correlated
#windspeedkmph and windGustKmph are positivly correlated
#WindChillF and tempF are positivly correlated
df1 = df1.drop(['Quarter','WindGustKmph','WindChillF','DepDelayMinutes'],axis=1)
```

```python
df1
```

```python
X = df1.drop('DelayClass',axis=1)
y = df1['DelayClass']
```

```python
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size=0.2)
```

```python
y_train.value_counts()
```

```python
sm = SMOTE(random_state=10)
X_train , y_train = sm.fit_resample(X_train , y_train)
```

```python
y_train.value_counts()
```

```python
model_LR = LogisticRegression(solver='lbfgs', max_iter=2000)
model_LR.fit(X_train,y_train)
```

```python
res = model_LR.predict(X_test)
```

```python
model_LR.score(X_test,y_test)
```

```python
print(classification_report(y_test,res))
```

```python
model_DTC = DecisionTreeClassifier()
model_DTC.fit(X_train,y_train)
```

```python
res = model_DTC.predict(X_test)
model_LR.score(X_test,y_test)
```

```python
print(classification_report(y_test,res))
```
