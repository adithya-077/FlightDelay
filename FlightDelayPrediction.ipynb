```python
# dataset file
os.chdir(cwd+'\Flight dataset')
lisofiles = os.listdir()
outerloop_reset = os.getcwd()
for i in lisofiles[:len(lisofiles)-1]:
    os.chdir(os.getcwd()+f'\\{i}')
    loopinit = os.getcwd()
    for i in os.listdir():
        print(i)
        temp = pd.DataFrame()
        os.chdir(os.getcwd()+f'\\{i}')
        print(os.getcwd())
        temp = pd.read_csv(os.listdir()[0],low_memory=False)
        df = pd.concat([temp,df])
        os.chdir(loopinit)
    os.chdir(outerloop_reset)
```

```python
# os.chdir('D:\\programming\\practice\\Project\\Weather dataset\\weather\\ATL')
# with open(os.getcwd()+f'\\2016-3.json') as f:
#     temp = json.load(f)
#     final = pd.DataFrame()
#     for i in range(len(temp['data']['weather'])):
#         date = temp['data']['weather'][i]['date']
#         for j in range(len(temp['data']['weather'][i]['hourly'])):
#             each = pd.DataFrame.from_dict(temp['data']['weather'][i]['hourly'][j])
#             each['date'] = date
#             final = pd.concat([each,final])
#             final = final.drop(['FeelsLikeF','winddir16Point','FeelsLikeC','DewPointC','windspeedMiles','HeatIndexF','HeatIndexC','weatherIconUrl','WindGustMiles','WindChillC','tempC','weatherDesc'],axis=1)
#         print(final)

#loading the weather datasets
weather = pd.DataFrame()
os.chdir(cwd)
os.chdir(cwd+'//Weather dataset//weather')
print(os.getcwd())
wloop_init = os.getcwd()
for i in os.listdir():
    airport = i
    os.chdir(os.getcwd()+f'//{i}')
    print(i)
    for j in os.listdir():
        if ('2016'==j[0:4]) or ('2017'==j[0:4]):
            print(j)
            with open(os.getcwd()+f'\\{j}') as f:
                temp = json.load(f)
                final = pd.DataFrame()
                for days in range(len(temp['data']['weather'])):
                    date = temp['data']['weather'][days]['date']
                    for hrs in range(len(temp['data']['weather'][days]['hourly'])):
                        each = pd.DataFrame.from_dict(temp['data']['weather'][days]['hourly'][hrs])
                        each['date'] = date
                        each['airport'] = airport
                        final = pd.concat([each,final])
                        final = final.drop(['FeelsLikeF','winddir16Point','FeelsLikeC','DewPointC','windspeedMiles','HeatIndexF','HeatIndexC','weatherIconUrl','WindGustMiles','WindChillC','tempC','weatherDesc'],axis=1)
                weather = pd.concat([weather,final],axis=0) 
                

    os.chdir(wloop_init)

weather
```
